```python
import data_functions
import pandas as pd

from matplotlib import pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
# Load data from a JSON Lines (JSONL) file
filtered_train_data = data_functions.load_jsonl('data/restaurant_filtered_train.jsonl')
filtered_test_data = data_functions.load_jsonl('data/restaurant_filtered_test.jsonl')
```

```python
train_counts_df, train_stacked_counts_df = data_functions.count_aspect_polarity(filtered_train_data)
train_stacked_counts_df
```

```python
test_counts_df, test_stacked_counts_df = data_functions.count_aspect_polarity(filtered_test_data)
test_stacked_counts_df
```

```python
plt.figure(figsize=(18, 6))
sns.barplot(data=train_stacked_counts_df, x="aspect", y="counts", hue="polarity")
plt.xticks(rotation=70, ha='right')
plt.show()
```

```python
plt.figure(figsize=(18, 6))
sns.barplot(data=test_stacked_counts_df, x="aspect", y="counts", hue="polarity")
plt.xticks(rotation=70, ha='right')
plt.show()
```

## Wordcloud

```python
train_text = ''
for d in filtered_train_data:
    train_text += d['text']
    train_text += ' '

    

wordcloud = WordCloud(width=1000, height=600, background_color="white", contour_color='steelblue', max_words=500).generate(train_text)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
test_text = ''
for d in filtered_test_data:
    test_text += d['text']
    test_text += ' '

    

wordcloud = WordCloud(width=1000, height=600, background_color="white", contour_color='steelblue', max_words=500).generate(test_text)

plt.figure(figsize=(10, 6))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
